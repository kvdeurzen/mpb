```python
import numpy as np
import pandas as pd
from pymongo import MongoClient
```

```python
client = MongoClient('mongodb://mp_test:mp_test@ds241395.mlab.com:41395/mp_baby_stuff')
db = client.mp_baby_stuff
collection = db.beds
```

```python
# Basic overview
print('Total number of items in DB: ', collection.count())

# Types
total= 0
for _type in collection.distinct('type'):
    count = collection.count({'type': _type})
    total = total + count
    print('Count in type', 
          _type, 
          ': ', 
          count)
print('Total: ', total)

# Categories
total = 0   
for _category in collection.distinct('category'):
    count = collection.count({'category': _category})
    total = total + count
    print('Count in category', 
          _category, 
          ': ', 
          count)
print('Total: ', total)

# Brands
total = 0
for _brand in collection.distinct('brand'):
    count = collection.count({'brand': _brand})
    total = total + count
    print('Count in brand ',
         _brand,
         ': ',
         count)
print('Total: ', total)

# Condition
total = 0
for _condition in collection.distinct('condition'):
    count = collection.count({'condition': _condition})
    total = total + count
    print('Count in condition ',
         _condition,
         ': ',
         count)
print('Total: ', total)
```
